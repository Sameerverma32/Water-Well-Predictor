```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
raw_path = 'groundwater_ml_dataset.csv'

state_names = [
    'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar',
    'Chhattisgarh', 'Goa', 'Gujarat', 'Haryana',
    'Himachal Pradesh', 'Jharkhand', 'Karnataka', 'Kerala',
    'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya',
    'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan',
    'Sikkim', 'Tamil Nadu', 'Telangana', 'Tripura',
    'Uttar Pradesh', 'Uttarakhand', 'West Bengal',
    'Chandigarh', 'Delhi', 'Jammu And Kashmir', 'Puducherry',
    'Haveli and Daman', 'Ladakh', 'Lakshadweep',
    'A & N ISLAND (UT)', 'PUDUCHERRY (UT)', 'LADAKH(UT)'
]

state_norm = {
    ''.join(ch for ch in name.upper() if ch.isalnum()): name
    for name in state_names
}

with open(raw_path, 'r', encoding='utf-8', errors='replace') as f:
    header = f.readline().strip().split(',')
    current_state = None
    rows = []
    for line in f:
        line = line.strip()
        if not line:
            continue
        if line.startswith('1,2,3.0,4.0,5.0,6.0,Safe'):
            continue
        parts = line.split(',')
        if len(parts) != len(header):
            continue

        raw_state = parts[1].strip()
        normalized = ''.join(ch for ch in raw_state.upper() if ch.isalnum())
        if normalized in state_norm:
            current_state = state_norm[normalized]
            continue

        if current_state is None:
            continue
        if not parts[0].strip().isdigit():
            continue

        rows.append([current_state, raw_state] + [p.strip() for p in parts[2:]])


df = pd.DataFrame(rows, columns=header).drop_duplicates().reset_index(drop=True)
df.head(10)
```
