```python
fig = px.pie(df,
            names="category",
            title="Groundwater Category Distribution")

fig.show()
```

```python
fig = px.box(df,
            x="state",
            y="annual_extraction",
            color="state",
            title="Distribution of Extraction by State")

fig.show()
```

```python
fig = px.histogram(df,
                x="annual_extraction",
                nbins=50, color="annual_extraction",
                title="Distribution of Groundwater Extraction")

fig.show()
```

```python
fig = px.line(df,
            x="year",
            y="annual_extraction",
            color="state",
            title="Yearly Groundwater Extraction Trend")

fig.show()
```

```python
fig = px.bar(df,
            x="state",
            y="annual_extraction",
            color="state",
            title="State-wise Annual Groundwater Extraction")

fig.show()
```

```python
pivot = df.pivot_table(values='risk_score',
                    index='state',
                    columns='stress_level',
                    aggfunc='mean')

fig = px.imshow(pivot,
                title="Heatmap of Stress Level by State")

fig.show()
```

```python
fig = px.scatter(df,
                x="annual_recharge",
                y="annual_extraction",
                color="stress_level",
                size="risk_score",
                title="Recharge vs Extraction Analysis")

fig.show()
```

```python
fig = px.choropleth(df,
                    locations="state",
                    locationmode="country names",
                    color="annual_extraction",
                    scope="asia",
                    title="India Groundwater Extraction Map")

fig.show()
```

```python
fig = px.ecdf(df, x="annual_extraction",
            title="ECDF of Extraction")
fig.show()
```

```python
fig = px.treemap(df, path=["state", "district"],
                values="annual_extraction",
                title="Treemap of Extraction")
fig.show()
```

```python
fig = px.sunburst(df, path=["state", "district", "category"],
                values="annual_extraction",
                title="Sunburst Chart")
fig.show()
```

```python
fig = px.funnel(df, x="annual_extraction", y="category",
                title="Funnel Chart")
fig.show()
```

```python
fig = px.box(df, x="category", y="annual_extraction",
            points="all", title="Box with Points")
fig.show()
```

```python
fig = px.violin(df, x="category", y="annual_extraction",
                box=True, title="Violin Plot with Box")
fig.show()
```
