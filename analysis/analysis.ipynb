```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('groundwater_dataset.csv')
df.head(10)
```

```python
fig = px.bar(df,
            x="state",
            y="annual_recharge",
            color="state",
            title="Annual Groundwater Recharge by State")

fig.show()
```

```python
fig = px.bar(df,
            x="district",
            y="annual_extraction",
            color="state",
            title="Annual Groundwater Extraction by District")

fig.show()
```

```python
fig = px.pie(df,
            names="category",
            title="Groundwater Category Distribution")

fig.show()
```

```python
fig = px.scatter(df,
                x="annual_recharge",
                y="annual_extraction",
                color="category",
                hover_data=["state","district"],
                title="Recharge vs Extraction")

fig.show()
```

```python
fig = px.box(df,
            x="category",
            y="stage_of_development",
            color="category",
            title="Stage of Development by Category")

fig.show()
```

```python
fig = px.histogram(df,
                x="stage_of_development",
                nbins=20,
                title="Distribution of Stage of Development")

fig.show()
```

```python
fig = px.line(df,
            x="district",
            y=["annual_recharge","annual_extraction"],
            title="Recharge vs Extraction Trend")

fig.show()
```
